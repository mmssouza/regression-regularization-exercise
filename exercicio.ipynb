# Exercício de regressão e regularização
### Os seis conjuntos de dados plotados foram produzidos com a rotina ```sklearn.datasets.make_regression()```.
### Aproxime funções com uma rede MLP para cada conjunto, dividindo as amostras em grupos de treino e validação, utilizando as de treino para regressão e de validação para avaliação de sobreajuste. Em cada caso, reporte as métricas de desempenho erro quadrático médio (MSE) e coeficiente de determinação (R²). Obtenha os resultados sem regularização e com as regularizações L1 e L2.

![Figura](fig1.png)


````python
import pickle
import numpy as np
import matplotlib.pyplot as plt

datasets = ['data'+str(i)+'.pkl' for i in range(1,7)]
data = []

for dataset in datasets:
 with open(dataset,"rb") as f:
  data.append(pickle.load(f))

fig, axs = plt.subplots(nrows= 2, ncols = 3) # 2 rows, 1 column

axs[0,0].plot(data[0][0],data[0][1],'.')
axs[0,1].plot(data[1][0],data[1][1],'.')
axs[0,2].plot(data[2][0],data[2][1],'.')
axs[1,0].plot(data[3][0],data[3][1],'.')
axs[1,1].plot(data[4][0],data[4][1],'.')
axs[1,2].plot(data[5][0],data[5][1],'.')
#axs[2,0].plot(data[6][0],data[6][1],'o')

# Adjust layout to prevent overlap
fig.suptitle("Conjuntos de dados para realização de exercício de regressão e regularização")
fig.supxlabel('x')
fig.supylabel('y')
plt.tight_layout()
#plt.savefig("fig1.png",dpi=100)
plt.show()
````
